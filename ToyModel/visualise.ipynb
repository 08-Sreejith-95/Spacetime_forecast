```python
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
#sys.path.append(os.path.dirname(os.path.dirname(os.path.abspath(__file__))))
print(os.getcwd())
```

```python
data = np.load('data.npz', allow_pickle= True)
train_samples = data['train']
train_time_steps = data['train_time']
val_samples = data['val']
val_time_steps = data['val_time']


ix = -1
plt.plot(train_samples[:, ix], label=f'(train)', alpha=1)
plt.plot(np.arange(len(train_samples), len(train_samples) + len(val_samples)), 
         val_samples[:, ix], label=f'(val)', alpha=1)
plt.legend()
plt.show()
```
